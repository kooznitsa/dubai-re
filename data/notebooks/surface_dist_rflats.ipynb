```python
import pandas as pd
import numpy as np
from ast import literal_eval
import re
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()
%matplotlib notebook
```

```python
df = pd.read_csv('../data/bayut/cleaned/ready_flats_clean.csv', converters={'amenities': literal_eval})
df.set_index('listing_id', inplace=True)
```

```python
sns.histplot(df[(df['surface'] <= 4000)]['surface'].astype(float), bins=30)
plt.xlabel('Площадь, кв. футов')
plt.ylabel('Количество')
plt.title('Распределение площадей вторичных квартир в Дубае', size=16)
```
